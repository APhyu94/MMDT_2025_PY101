```python
import csv

def grams_to_myanmar_units(grams):
    kyat = int(grams // 16.3293)
    remaining = grams % 16.3293
    pae = int(remaining // 4.0823)
    remaining %= 4.0823
    yway = round(remaining / 0.40823)
    return kyat, pae, yway

def myanmar_units_to_grams(kyat, pae, yway):
    return kyat * 16.3293 + pae * 4.0823 + yway * 0.40823

with open("/content/drive/MyDrive/Gold_Sales.csv", "r") as file:
    reader = csv.DictReader(file)
    for row in reader:
        # Optional: Clean up keys/values
        row = {key.strip(): value.strip() for key, value in row.items()}

        item = row["Item"]
        price_per_gram = float(row["Price per gram"].replace(',', ''))  # <-- fixed here
        grams = float(row["Grams"])
        buyer = row["Buyer"]
        date = row["Date"]
        com_kyat = float(row["ComKyat"])
        com_pae = float(row["ComPae"])
        com_yway = float(row["ComYway"])

        kyat, pae, yway = grams_to_myanmar_units(grams)
        total_price = grams * price_per_gram
        commission_grams = myanmar_units_to_grams(com_kyat, com_pae, com_yway)
        commission_value = commission_grams * price_per_gram
        total_value_with_commission = total_price + commission_value

        print("==== Sales Voucher ====")
        print(f"Date: {date}")
        print(f"Buyer: {buyer}")
        print(f"Item: {item}")
        print(f"Gold Weight: {kyat} ကျပ် {pae} ပဲ {yway} ရွေး ({grams:.2f} grams)")
        print(f"Price per gram: {price_per_gram:,.0f} Ks")
        print(f"Total Gold Price: {total_price:,.0f} Ks")
        print(f"Commission Taken (အလျော့တွက်): {com_kyat} ကျပ် {com_pae} ပဲ {com_yway} ရွေး")
        print(f"Commission Value in Ks: {commission_value:,.0f} Ks")
        print(f"Total Value of Item (including commission): {total_value_with_commission:,.0f} Ks")
        print("=======================\n")
```
